```python
import pandas as pd
import matplotlib.pyplot as plt

# ------------------------------------------------------------
# 1. Load Dataset (Kaggle path)
# ------------------------------------------------------------
df = pd.read_csv("/kaggle/input/students-score/StudentsPerformance.csv")

# ------------------------------------------------------------
# 2. Data Validation
# ------------------------------------------------------------
df = df.dropna()

for col in ["math score", "reading score", "writing score"]:
    df[col] = pd.to_numeric(df[col], errors="coerce")
    df = df[df[col] >= 0]

# ------------------------------------------------------------
# 3. Calculate subject averages
# ------------------------------------------------------------
avg_math = df["math score"].mean()
avg_reading = df["reading score"].mean()
avg_writing = df["writing score"].mean()

print("\n--- Subject Averages ---")
print("Math Average     :", avg_math)
print("Reading Average  :", avg_reading)
print("Writing Average  :", avg_writing)

# ------------------------------------------------------------
# 4. Top 5 performers
# ------------------------------------------------------------
df["total_score"] = df["math score"] + df["reading score"] + df["writing score"]
top5 = df.sort_values(by="total_score", ascending=False).head(5)

print("\n--- Top 5 Students ---")
print(top5[["gender","race/ethnicity","math score","reading score","writing score","total_score"]])

# ------------------------------------------------------------
# 5. Visualization
# ------------------------------------------------------------
plt.figure(figsize=(7,5))
plt.bar(["Math", "Reading", "Writing"], [avg_math, avg_reading, avg_writing])
plt.title("Average Subject Scores")
plt.xlabel("Subjects")
plt.ylabel("Average Marks")
plt.tight_layout()
plt.show()





```
